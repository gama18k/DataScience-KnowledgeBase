# Matplotlib

## Gr√°fico de linha

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 35]

plt.plot(x, y, label="Linha de Exemplo", color="blue", marker="o")
plt.title("Gr√°fico de Linha")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.legend()
plt.grid(True)
plt.show()
```

## Gr√°fico de disper√ß√£o

```python
x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]

plt.scatter(x, y, color="red")
plt.title("Gr√°fico de Dispers√£o")
plt.xlabel("Valores X")
plt.ylabel("Valores Y")
plt.show()
```

## Histograma

```python
data = [7, 8, 5, 6, 7, 8, 8, 10, 11, 12, 7, 6, 5, 4, 5]

plt.hist(data, bins=5, color="green", edgecolor="black")
plt.title("Histograma")
plt.xlabel("Intervalos")
plt.ylabel("Frequ√™ncia")
plt.show()
```

## Barras horizontais

```python
categorias = ["A", "B", "C", "D"]
valores = [3, 7, 8, 5]
plt.barh(categorias, valores, color="green")
plt.title("Gr√°fico de Barras Horizontais")
plt.xlabel("Valores")
plt.ylabel("Categorias")
plt.show()
```

## Subplots

```python

x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [2, 4, 6, 8]

plt.subplot(1, 2, 1)  # 1 linha, 2 colunas, 1¬∫ gr√°fico
plt.plot(x, y1, color="blue")
plt.title("Gr√°fico 1")

plt.subplot(1, 2, 2)  # 1 linha, 2 colunas, 2¬∫ gr√°fico
plt.plot(x, y2, color="green")
plt.title("Gr√°fico 2")

plt.tight_layout()  # Ajusta os espa√ßos entre os gr√°ficos
plt.show()
```

## Subplot personalizado

```python
plt.style.use("ggplot")  # Exemplos: 'seaborn', 'classic', etc.
plt.plot([1, 2, 3], [1, 4, 9], label="Curva")
plt.legend()
plt.show()
```

## Pizza

```python
sizes = [15, 30, 45, 10]
labels = ["A", "B", "C", "D"]

plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=140)
plt.title("Gr√°fico de Pizza")
plt.show()
```

## Gr√°fico com preenchimento

```python
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

plt.fill_between(x, y, color="skyblue", alpha=0.4)
plt.plot(x, y, color="blue")
plt.title("Gr√°fico com Preenchimento")
plt.show()
```
