# Seaborn

## Gráfico de dispersão

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame({
    "x": [5, 7, 8, 7, 2, 17, 2, 9, 4, 11],
    "y": [99, 86, 87, 88, 100, 86, 103, 87, 94, 78],
    "categoria": ["A", "B", "A", "B", "A", "B", "A", "B", "A", "B"]
})

sns.scatterplot(data=data, x="x", y="y", hue="categoria", style="categoria", palette="coolwarm")
plt.title("Gráfico de Dispersão com Hue e Style")
plt.show()
```

## Gráfico de linhas

```python
data = pd.DataFrame({
    "x": [5, 7, 8, 7, 2, 17, 2, 9, 4, 11],
    "y": [99, 86, 87, 88, 100, 86, 103, 87, 94, 78],
    "categoria": ["A", "B", "A", "B", "A", "B", "A", "B", "A", "B"]
})

sns.lineplot(x="x", y="y", data=data)
plt.title("Gráfico de Linha")
plt.show()
```

## Boxplot

```python
data = pd.DataFrame({
    "categoria": ["A", "B", "A", "B", "A", "B", "A", "B", "A", "B"],
    "valores": [10, 20, 15, 25, 10, 30, 15, 25, 20, 30]
})

sns.boxplot(data=data, x="categoria", y="valores", palette="Set2")
plt.title("Boxplot por Categoria")
plt.show()
```

## Barplot

```python
data = pd.DataFrame({
    "categoria": ["A", "B", "A", "B", "A", "B", "A", "B", "A", "B"],
    "valores": [10, 20, 15, 25, 10, 30, 15, 25, 20, 30]
})

sns.barplot(x="categoria", y="valores", data=data, ci="sd", palette="muted")
plt.title("Barplot")
plt.show()
```

## Catplot

```python
# Carregar o dataset Tips
tips = sns.load_dataset('tips')

# Strip Plot
sns.catplot(data=tips, x='day', y='total_bill', kind='strip', hue='sex', palette='Set2')
plt.title('Catplot - Strip Plot')
plt.show()

# Swarm Plot
sns.catplot(data=tips, x='day', y='total_bill', kind='swarm', hue='sex', palette='Set2')
plt.title('Catplot - Swarm Plot')
plt.show()

# Box Plot
sns.catplot(data=tips, x='day', y='total_bill', kind='box', hue='sex', palette='Set2')
plt.title('Catplot - Box Plot')
plt.show()

# Violin Plot
sns.catplot(data=tips, x='day', y='total_bill', kind='violin', hue='sex', split=True, palette='Set2')
plt.title('Catplot - Violin Plot')
plt.show()

# Boxen Plot
sns.catplot(data=tips, x='day', y='total_bill', kind='boxen', hue='sex', palette='Set2')
plt.title('Catplot - Boxen Plot')
plt.show()

# Point Plot
sns.catplot(data=tips, x='day', y='total_bill', kind='point', hue='sex', palette='Set2')
plt.title('Catplot - Point Plot')
plt.show()

# Bar Plot
sns.catplot(data=tips, x='day', y='total_bill', kind='bar', hue='sex', palette='Set2', ci='sd')
plt.title('Catplot - Bar Plot')
plt.show()

# Count Plot
sns.catplot(data=tips, x='day', kind='count', hue='sex', palette='Set2')
plt.title('Catplot - Count Plot')
plt.show()
```

## Violin plot

```python
iris = sns.load_dataset('iris')

plt.figure(figsize=(8, 6))
sns.violinplot(x='species', y='sepal_length', data=iris, palette='muted')
plt.title('Violin Plot de sepal_length por species')
plt.show()
```

## Pairplot

```python
sns.pairplot(iris, hue='species', palette='muted')
plt.suptitle('Pairplot do dataset Iris', y=1.02)
plt.show()
```

## Mapa de calor

```python
import numpy as np

# Dados de exemplo
matrix = np.random.rand(5, 5)

sns.heatmap(matrix, annot=True, cmap="viridis")
plt.title("Mapa de Calor")
plt.show()
```

## Histograma com KDE

```python

data = [7, 8, 5, 6, 7, 8, 8, 10, 11, 12, 7, 6, 5, 4, 5]

sns.histplot(data, kde=True, color="purple")
plt.title("Histograma com KDE", fontsize=15)
plt.xlabel("Valores")
plt.ylabel("Frequência")
plt.show()
```
